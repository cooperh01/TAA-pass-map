```python
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer.pitch import Pitch
import seaborn as sns
```

```python
#Read in the data
df = pd.read_csv('Champions-League-Finals-19.csv')

df = df[(df['player_name'] == 'Trent Alexander-Arnold')]
df = df[(df['event_type'] == 'Pass')]
df = df.reset_index()

df[['player_name', 'event_type', 'pass_outcome_name','location_0','location_1', 'pass_end_location_0', 'pass_end_location_1']]
```

```python
fig ,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')

#Pitch Template
pitch = Pitch(pitch_type='statsbomb', orientation='horizontal',
              pitch_color='#22312b', line_color='#c7d5cc', figsize=(13, 8),
              constrained_layout=False, tight_layout=True)

pitch.draw(ax=ax)


#Plot each pass
for x in range(len(df['location_0'])):
    if df['pass_outcome_name'][x] == 'Complete':
        plt.plot((df['location_0'][x],df['pass_end_location_0'][x]),(df['location_1'][x],df['pass_end_location_1'][x]),color='green')
        plt.scatter(df['location_0'][x],df['location_1'][x],color='green')
    if df['pass_outcome_name'][x] == 'Incomplete':
        plt.plot((df['location_0'][x],df['pass_end_location_0'][x]),(df['location_1'][x],df['pass_end_location_1'][x]),color='orange')
        plt.scatter(df['location_0'][x],df['location_1'][x],color='orange')
    if df['pass_outcome_name'][x] == 'Out':
        plt.plot((df['location_0'][x],df['pass_end_location_0'][x]),(df['location_1'][x],df['pass_end_location_1'][x]),color='red')
        plt.scatter(df['location_0'][x],df['location_1'][x],color='red')

plt.title('Trent Alexander-Arnold Pass Map 2019 Champions League Final',color='white',size=20)
plt.figtext(.3,.86,'Successfull',fontsize=14,ha='left',color='green')
plt.figtext(.45,.86,'Unsuccessfull',fontsize=14,ha='left',color='orange')
plt.figtext(.65,.86,'Out',fontsize=14,ha='left',color='red')


plt.show()
```
